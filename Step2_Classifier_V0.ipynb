```python
def train(model_name, load_weights_path):
    
    batch_size = 32

    for i in range(10):

        model_name_i =model_name + "_" + str(i+1)

        #获得训练和测试集合，以：路径、class label的形式保存
        train_files, holdout_files = get_train_holdout_files( current_iteration = (i+1) )
        
        #训练数据集
        train_gen = data_generator(batch_size, train_files, train_set=True)

        #测试数据集
        val_x,val_y = get_valid_data(holdout_files)

        #动态设置学习率
        learnrate_scheduler = LearningRateScheduler(step_decay)

        #获取model
        model = get_net(load_weight_path=load_weights_path)

        checkpoint = ModelCheckpoint("workdir/model_" + model_name_i + "_"  + "_e" + "{epoch:02d}-{val_loss:.4f}.hd5", monitor='val_loss', verbose=1, save_best_only=False, save_weights_only=False, mode='auto', period=1)

        checkpoint_best = ModelCheckpoint("workdir/model_" + model_name_i + "_"  + "_best.hd5", monitor='val_loss', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', period=1)

        metrics = Metrics()
        
        history = model.fit_generator(generator=train_gen, samples_per_epoch=len(train_files) / 1, validation_data=(val_x,val_y), nb_epoch=10, verbose=1, class_weight="auto", callbacks=[checkpoint, checkpoint_best, learnrate_scheduler, metrics])

        model.save("workdir/model_" + model_name_i + "_end.hd5")
```

```python
def step_decay(epoch):
    res = 0.001
    if epoch > 100:
        res = 0.0001
    print("learnrate: ", res, " epoch: ", epoch)
    return res
```

```python
class Metrics(Callback):  
        
    def on_epoch_end(self, epoch, logs=None):       
        
        logs = logs or {}        
        val_predict = np.argmax(self.model.predict(self.validation_data[0]), -1)     
        val_targ = self.validation_data[1]     
        
        #是否是one-hot编码的多分类
        if len(val_targ.shape) == 2 and val_targ.shape[1] != 1: 
            val_targ = np.argmax(val_targ, -1) 

        val_f1 = f1_score(val_targ, val_predict, average='macro')        
        val_recall = recall_score(val_targ, val_predict, average='macro')        
        val_precision = precision_score(val_targ, val_predict, average='macro')     

        logs['val_f1'] = val_f1        
        logs['val_recall'] = val_recall        
        logs['val_precision'] = val_precision    

        print(" — val_f1: %f — val_precision: %f — val_recall: %f" % (val_f1, val_precision, val_recall))
 
```

```python
def get_net(input_shape=(Height, Weight, Channel), load_weight_path=None) -> Model:  #期待返回类型为model
    
    inputs = Input(shape=input_shape, name="input")
    x = inputs
    
    ##################################################################################################################
    x_ident_1 = x
    x_ident_1 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x_ident_1)
    # 1st layer group
    x = Convolution2D(64, 3, 3, activation=None, border_mode='same', name='conv1a', subsample=(1, 1))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Convolution2D(64, 3, 3, activation=None, border_mode='same', name='conv1b', subsample=(1, 1))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid', name='pool1')(x)
    x_ident_2 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x)
    x = Concatenate(axis=3)([x,x_ident_1])
    
    weight = layers.GlobalAveragePooling2D(name='weight_1')(x)  
    weight = layers.Reshape((1, 1, 67), name='weight_reshape_1')(weight)
    weight = layers.Conv2D(67, 1, padding='same', activation='sigmoid', name='weight_1_reduce')(weight)
    if USE_DROPOUT:
        weight = Dropout(p=0.5)(weight)
    x = layers.multiply([x, weight], name='weight_1_excite')
    x = BatchNormalization()(x)
    
    ##################################################################################################################
    x_ident_1 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x_ident_1)
    # 2nd layer group
    x = Convolution2D(128, 3, 3, activation=None, border_mode='same', name='conv2a', subsample=(1, 1))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Convolution2D(128, 3, 3, activation=None, border_mode='same', name='conv2b', subsample=(1, 1))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid', name='pool2')(x)
    x_ident_3 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x)
    x = Concatenate(axis=3)([x,x_ident_1,x_ident_2])
    
    weight = layers.GlobalAveragePooling2D(name='weight_2')(x)  
    weight = layers.Reshape((1, 1, 195), name='weight_reshape_2')(weight)
    weight = layers.Conv2D(195, 1, padding='same', activation='sigmoid', name='weight_2_reduce')(weight)
    if USE_DROPOUT:
        weight = Dropout(p=0.5)(weight)
    x = layers.multiply([x, weight], name='weight_2_excite')
    x = BatchNormalization()(x)

    ##################################################################################################################
    x_ident_1 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x_ident_1)
    x_ident_2 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x_ident_2)
    # 3rd layer group
    x = Convolution2D(256, 3, 3, activation=None, border_mode='same', name='conv3a', subsample=(1, 1))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Convolution2D(256, 3, 3, activation=None, border_mode='same', name='conv3b', subsample=(1, 1))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid', name='pool3')(x)
    x_ident_4 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x)
    x = Concatenate(axis=3)([x,x_ident_1,x_ident_2,x_ident_3])
    
    weight = layers.GlobalAveragePooling2D(name='weight_3')(x)  
    weight = layers.Reshape((1, 1, 451), name='weight_reshape_3')(weight)
    weight = layers.Conv2D(451, 1, padding='same', activation='sigmoid', name='weight_3_reduce')(weight)
    if USE_DROPOUT:
        weight = Dropout(p=0.5)(weight)
    x = layers.multiply([x, weight], name='weight_3_excite')
    x = BatchNormalization()(x)
     
    ##################################################################################################################
    x_ident_1 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x_ident_1)
    x_ident_2 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x_ident_2)
    x_ident_3 = AveragePooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid')(x_ident_3)
   
    # 4th layer group
    x = Convolution2D(512, 3, 3, activation=None, border_mode='same', name='conv4a', subsample=(1, 1),)(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Convolution2D(512, 3, 3, activation=None, border_mode='same', name='conv4b', subsample=(1, 1),)(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), border_mode='valid', name='pool4')(x)
    x = Concatenate(axis=3)([x,x_ident_1,x_ident_2,x_ident_3,x_ident_4])
    
    weight = layers.GlobalAveragePooling2D(name='weight_4')(x)  
    weight = layers.Reshape((1, 1, 963), name='weight_reshape_4')(weight)
    weight = layers.Conv2D(963, 1, padding='same', activation='sigmoid', name='weight_4_reduce')(weight)
    if USE_DROPOUT:
        weight = Dropout(p=0.5)(weight)
    x = layers.multiply([x, weight], name='weight_4_excite')
    x = BatchNormalization()(x)
     
    
    ##################################################################################################################
    x = GlobalMaxPooling2D()(x)
    x = BatchNormalization(name="final_features")(x)
    
    ##################################################################################################################
    if USE_DROPOUT:
        x = Dropout(p=0.5)(x)
        
    x = Dense(128, activation='relu')(x)
    out_class = Dense(4, activation='softmax', name='out_class')(x)

    model = Model(input=inputs, output=out_class)
    
    if load_weight_path is not None:
        model.load_weights(load_weight_path, by_name=False)

    #编译模型
    model.compile(optimizer=SGD(lr=LEARN_RATE, momentum=0.9, nesterov=True), loss={ "out_class": "categorical_crossentropy" }, metrics={"out_class": [categorical_accuracy, categorical_crossentropy] } ) 
    model.summary(line_length=120)

    return model
```

```python
def data_generator(batch_size, record_list, train_set):
    
    while True:
        
        batch_index = 0
        image_list = []
        label_list = []

        if train_set:
            random.shuffle(record_list)

        #逐一遍历所有数据
        for index, record_item in enumerate(record_list):

            sample_path = record_item[0]
            sample_label = record_item[1]

            #转换成多分类标签
            sample_label = np_utils.to_categorical(sample_label,4)  

            #读取图片、标准化、修改格式
            sample_image = cv2.imread(sample_path)
            sample_image = (sample_image - np.average(sample_image)) / np.std(sample_image)
            sample_image = sample_image.reshape(1, sample_image.shape[0], sample_image.shape[1], 3)

            #数据增强
            if train_set:  
                if random.randint(0, 100) > 50:
                    sample_image = np.fliplr(sample_image)
                if random.randint(0, 100) > 50:
                    sample_image = np.flipud(sample_image)
                if random.randint(0, 100) > 50:
                    sample_image = sample_image[:,::-1]
                if random.randint(0, 100) > 50:
                    sample_image = sample_image[::-1, :]

            #添加数据
            image_list.append(sample_image)
            label_list.append(sample_label)
            batch_index += 1

            if batch_index >= batch_size:
                x = np.vstack(image_list)
                y = np.vstack(label_list)
                yield x, y
                image_list = []
                label_list = []
                batch_index = 0
```

```python
def get_valid_data(record_list):
    
    image_list = []
    label_list = []
    
    #逐一遍历所有数据
    for index, record_item in enumerate(record_list):

        sample_path = record_item[0]
        sample_label = record_item[1]

        #转换成多分类标签
        sample_label = np_utils.to_categorical(sample_label,4)  

        #读取图片、标准化、修改格式
        sample_image = cv2.imread(sample_path)
        sample_image = (sample_image - np.average(sample_image)) / np.std(sample_image)
        sample_image = sample_image.reshape(1, sample_image.shape[0], sample_image.shape[1], 3)

        #添加数据
        image_list.append(sample_image)
        label_list.append(sample_label)

    x = np.vstack(image_list)
    y = np.vstack(label_list)
    
    return x, y
```

```python
def get_train_holdout_files(current_iteration = 1):
    
    ############################读取训练和测试的样本###############################
    src_dir = Base_Path + "Data_Description/" + str(current_iteration) +"/"
    train_samples = pd.read_csv(src_dir + "train.csv").values
    holdout_samples = pd.read_csv(src_dir + "holdout.csv").values
    
    print("Train Count: ", len(train_samples), ", Holdout Count: ", len(holdout_samples))
    
    ##############################训练集数据平衡###################################
    train_samples_x = train_samples[:,0].reshape(-1,1)
    train_samples_y = train_samples[:,1].reshape(-1,1)
    

    ros = over_sampling.RandomOverSampler(random_state=0)
    train_samples_x, train_samples_y = ros.fit_sample(train_samples_x, train_samples_y.astype('int'))
    print("Train Sample_X: ", len(train_samples_x), ", Train Sample_Y: ", len(train_samples_y))

    sns.countplot(train_samples_y)

    train_samples_x = train_samples_x.reshape(-1,1)
    train_samples_y = train_samples_y.reshape(-1,1)
    train_samples = np.hstack((train_samples_x, train_samples_y)).tolist()
    
    print("Train Count: ", len(train_samples), ", Holdout Count: ", len(holdout_samples))
    random.shuffle(train_samples)
    print("Train Count: ", len(train_samples), ", Holdout Count: ", len(holdout_samples))
    
    return train_samples, holdout_samples
```
